```python
import pandas as pd

# Create a sample dataframe with a timedelta64[ns] column
df = pd.DataFrame({
    'Start Time': pd.to_datetime(['2022-01-01 12:00:00', '2022-01-01 13:00:00']),
    'End Time': pd.to_datetime(['2022-01-01 13:00:00', '2022-01-01 15:00:00'])
})

# Calculate the duration of each row
df['Duration'] = df['End Time'] - df['Start Time']

# Calculate the sum of the Duration column
total_duration = df['Duration'].sum()

print(total_duration)
```

```python
# Group data by "Project", "Point: Code", "Date" and "Исполнитель" columns
grouped_data = df.groupby(['Исполнитель','Проекты', 'Точка: Customer Name', 'Дата'])

# Calculate the difference between the maximum and minimum "End Time" values for each group
result = grouped_data.agg({'Дата окончания работы': lambda x: x.max() - x.min()})

# Rename the "End Time" column to "Working Time"
result = result.rename(columns={'Дата окончания работы': 'Working Time'})

# Group the result by "Artist" and calculate the total work time for each performer
total_work_time = result.groupby('Исполнитель')['Working Time'].sum()

print(total_work_time)
```
